```python
# Normal (Gaussian) Distribution

# The normal distribution is a continuous probability distribution characterized by its mean (μ) and standard deviation (σ).  
# It is widely used in statistics and natural sciences.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate 1000 samples from a normal distribution (mean=0, std=1)
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot with seaborn
sns.histplot(data, bins=30, kde=True, color="skyblue")

plt.title("Normal Distribution (mean=0, std=1)")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()
```
