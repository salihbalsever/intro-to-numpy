```python
# Zipf Distribution

# The Zipf distribution is a discrete power-law distribution.  
# It models data where a few items are very common, and many items are rare â€” like word frequencies in a language or website traffic.

# It has two parameters:
# - `a` (exponent): controls the distribution steepness  
# - `size`: number of random values to generate
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate 1000 samples from a Zipf distribution with exponent a = 2
data = np.random.zipf(a=2, size=1000)

# Limit extreme values for clearer visualization
filtered_data = data[data < 10]

# Plot with seaborn
sns.histplot(filtered_data, bins=range(1, 11), discrete=True, color="slateblue", edgecolor="black")

plt.title("Zipf Distribution (a = 2)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()

# Generates 1000 values from a Zipf distribution  
# with exponent `a = 2`.

# - Lower values occur much more frequently than higher values  
# - Commonly used in natural language processing, web traffic, and social networks
```
