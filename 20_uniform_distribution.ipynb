```python
# Uniform Distribution

# A uniform distribution has constant probability across a given range.  
# Every value between the minimum and maximum is equally likely.

# In NumPy, we use `np.random.uniform(low, high, size)` to generate uniform data.
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Generate 1000 samples between 10 and 20
data = np.random.uniform(low=10, high=20, size=1000)

# Plot with seaborn
sns.histplot(data, bins=30, color="cornflowerblue", edgecolor="black")

plt.title("Uniform Distribution (10 to 20)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
