```python
# Chi-Square Distribution

# The chi-square distribution is a special case of the gamma distribution and is commonly used in hypothesis testing (like the Chi-Square test for independence).

# It is defined by:
# - `df` (degrees of freedom): the number of independent variables
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate 1000 chi-square distributed samples with df=4
data = np.random.chisquare(df=4, size=1000)

# Plot with seaborn
sns.histplot(data, bins=30, kde=True, color="goldenrod", edgecolor="black")

plt.title("Chi-Square Distribution (df = 4)")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()

#Generates 1000 samples from a chi-square distribution  
#with 4 degrees of freedom (df = 4).
```
