```python
# Exponential Distribution

# The exponential distribution models the time between events in a Poisson process.  
# It has a single parameter:
# - `scale` = 1 / λ → the inverse of the event rate

# In NumPy:
# `np.random.exponential(scale, size)`
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate 1000 samples from an exponential distribution with scale = 2.0
data = np.random.exponential(scale=2.0, size=1000)

# Plot with seaborn
sns.histplot(data, bins=30, kde=True, color="tomato", edgecolor="black")

plt.title("Exponential Distribution (scale = 2.0)")
plt.xlabel("Time between Events")
plt.ylabel("Density")

plt.show()
```
