```python
# Poisson Distribution

# The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space.

# It is defined by a single parameter:
# - `Î»` (lambda): the expected number of occurrences within the interval
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Generate 1000 samples from a Poisson distribution with Î»=5
# lam=5 â expected number of events
# size=1000 â how many values to generate

data = np.random.poisson(lam=5, size=1000)

# Plot with seaborn
sns.histplot(data)

plt.title("Poisson Distribution (Î» = 5)")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")

plt.show()
```
