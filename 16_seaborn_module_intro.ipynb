```python
# Seaborn Module Introduction

# Seaborn is a Python data visualization library built on top of Matplotlib. It makes it easy to create attractive and informative statistical graphics.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.displot([0, 1, 2, 3, 4, 5])

plt.show()
```

```python
sns.displot([0, 1, 2, 3, 4, 5], kind="kde")

plt.show()
```

```python
# Load example dataset
tips = sns.load_dataset("tips")
```

```python
# Create a box plot showing total bill by day
sns.boxplot(x="day", y="total_bill", data=tips)

plt.show()
```
