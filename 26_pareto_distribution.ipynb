```python
# Pareto Distribution

# The Pareto distribution is a power-law probability distribution used to describe many real-world phenomena  
# where a small number of events account for most of the effect (e.g., wealth distribution).

# It has two key parameters:
# - `a` (shape): controls the steepness of the distribution  
# - `size`: number of random values to generate
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate 1000 samples from a Pareto distribution with shape parameter a = 3
data = np.random.pareto(a=3.0, size=1000)

# Shift values to make plot more readable (Pareto is defined for x > 1)
data = data + 1

# Plot with seaborn
sns.histplot(data, bins=30, kde=True, color="darkorange", edgecolor="black")

plt.title("Pareto Distribution (a = 3.0)")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()

# Generates 1000 values from a Pareto distribution  
# with shape parameter `a = 3.0`.

# - `a`: higher values make the tail shorter (steeper drop)  
# - `size`: number of random samples
```
